```python
#실습2: 최대값 구하기
number = input("numbers? ")
num = number.split()
n = len(number.split())
for i in range(0, n):
    for j in range(1, n):
        if int(num[i])<int(num[j]) :
            temp = num[i]
            num[i] = num[j]
            num[j] = temp
print("Max = ", num[0])
```

```python
#실습3: 암호화/복호화
def encoding(msg):
    plus = {}
    answer = ''
    
    for i in range(len(msg)):
        if msg[i]==' ':
            plus[i] = msg[i]
            continue
        plus[i] = chr(ord(msg[i])+1)
    
    for i in range(len(msg)):
        if plus[i]==' ':
            for j in range(i+1,len(msg),1):
                answer = answer+plus[j]
            answer = answer+' '
            for k in range(0,len(msg)-i-1,1):
                answer = answer+plus[k]
        
    return answer

def decoding(msg):
    back = {}
    answer = ''
    
    for i in range(len(msg)):
        if msg[i]==' ':
            back[i] = msg[i]
            continue
        back[i] = chr(ord(msg[i])-1)
    
    for i in range(len(msg)):
        if back[i]==' ':
            for j in range(i+1,len(msg),1):
                answer = answer+back[j]
            answer = answer+' '
            for k in range(0,len(msg)-i-1,1):
                answer = answer+back[k]
        
    return answer

def main():
    msg = input()
    enc = encoding(msg)
    dec = decoding(enc)
    print(msg+" -> 암호화 -> "+enc)
    print(enc+" -> 복호화 -> "+dec)
    
main()
```

```python
#실습4: 전화번호부
def print_menu():
    print("1. 연락처 입력")
    print("2. 연락처 출력")
    print("3. 연락처 이름검색")
    print("4. 연락처 번호검색")
    print("5. 종료")
    menu = input("메뉴선택 : ")
    print("\n")
    return int(menu)

def main():
    info = []
    name = []
    num = []
    while 1:
        menu = print_menu()
        if menu == 1:
            info = enter()
            name = name + info[0]
            num = num + info[1]
        elif menu == 2:
            print("*** 전화번호부 ***")
            for i in range(len(name)):
                print(name[i-1]+" "+num[i-1])
            print("\n")
        elif menu == 3:
            n = input("이름: ")
            for i in range(len(name)):
                if n == name[i-1]:
                    print("검색결과 >> "+n+" "+num[i-1])
                    print("\n")
        elif menu == 4:
            p = input("전화번호: ")
            for i in range(len(num)):
                if p == num[i-1]:
                    print("검색결과 >> "+name[i-1]+" "+p)
                    print("\n")
        elif menu == 5:
            break
        
def enter():
    print("*** 새 연락처 추가하기 ***")
    print("이름 전화번호 순으로 입력하기 --- 예) aaa 123")
    print("입력완료 시 z를 입력해주세요...")
    list1 = []
    list2 = []
    while 1:
        put = input()
        p = put.split()
        if put!= 'z':
            list1.append(p[0])
            list2.append(p[1])
        elif put == 'z':
            print("\n")
            break
    
    return list1, list2
    
if __name__ == "__main__":
    main()
```
