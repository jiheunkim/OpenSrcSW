```python
#실습1: 생일 기온 시각화
#지난 실습 csv파일과 구별하기 위해 seoul1.csv로 파일명
import csv
import matplotlib.pyplot as plt
f = open('seoul1.csv', 'r', encoding = 'cp949')
data = csv.reader(f)
next(data)
high = []
low = []
year = []

for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] == '05' and row[0].split('-')[2] == '04':
            high.append(float(row[-1]))
            year.append(row[0].split('-')[0])
    if row[-2] != '':
        if row[0].split('-')[1] == '05' and row[0].split('-')[2] == '04':
            low.append(float(row[-2]))
f.close()
plt.rc('font', family = 'Malgun Gothic')
plt.title('생일 기온 시각화')
plt.plot(year, high, color = 'mediumvioletred')
plt.plot(year, low, color = 'skyblue')
plt.show()
```

```python
#실습2: 최고 기온 시각화
import csv
import matplotlib.pyplot as plt
f = open('seoul1.csv', 'r', encoding = 'cp949')
data = csv.reader(f)
next(data)
jan = []
aug = []

for row in data:
    month = row[0].split('-')[1]
    if row[-1] != '':
        if month == '01':
            jan.append(float(row[-1]))
        if month == '08':
            aug.append(float(row[-1]))
plt.rc('font', family = 'Malgun Gothic')
plt.title('서울 지역 1,8월 최고 기온 시각화')
plt.hist(jan, bins = 100, color = 'b', label = 'Jan')
plt.hist(aug, bins = 100, color = 'r', label = 'Aug')
plt.legend()
plt.show()
```
